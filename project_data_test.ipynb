```python
import pickle
import pandas as pd
```

```python
new_dataset = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/ironkaggle_notarget.csv")
new_dataset.head()
```

```python
new_dataset[['Year', 'Month', 'Day']] = new_dataset['Date'].str.split('-', expand=True).astype(int)
new_dataset = new_dataset.drop(columns=['Date']) 
new_dataset.head()
```

```python
new_dataset['State_holiday'] = new_dataset['State_holiday'].apply(lambda x: 0 if x == '0' else 1).astype(int)
new_dataset = new_dataset.drop(columns = ["Store_ID", "True_index"])
new_dataset.head()
```

```python
model = pickle.load(open("gb_reg_grid_model.pkl", "rb"))
normalizer = pickle.load(open("normalizer.pkl", "rb"))
```

```python
df_norm = normalizer.transform(new_dataset)
df_norm = pd.DataFrame(df_norm, columns = new_dataset.columns)


new_dataset["pred_sales"] = model.predict(df_norm)
```
